# Análisis del dataframe

---

Procedemos a realizar un pequeño ajuste en la columna 'PlayerMarketValue'

### Transformación

---

Cambiamos el nombre de la columna 'PlayerMarketValue' al nuevo nombre llamado 'PlayerMarketValue(M/€)' y eliminamos el símbolo de '€' y la palabra 'million' de las filas. Esto con la finalidad de poder manipular los datos y realizar operaciones en Tableau.

```python
valores_convertidos = []

for valor in data['PlayerMarketValue']:
    valor_convertido = int(valor.replace('€', '').replace(' million', ''))
    
    valores_convertidos.append(valor_convertido)        # agrega el valor convertido a la lista

data['PlayerMarketValue(M/€)'] = valores_convertidos        # crea una nueva columna con los valores convertidos
data.drop('PlayerMarketValue', axis = 1, inplace = True)

data.head()
```

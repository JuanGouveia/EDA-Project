```python
import pandas as pd
```

```python
data = pd.read_csv('../../Data-Projects/data_visualizacion/footballbootsdb.csv')
```

```python
data.drop_duplicates().shape == data.shape
```

```python
mask = data.isnull()        # contamos los nulos

null_counts = mask.sum()

print(null_counts)
```

```python
data.drop(['Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17'], axis = 1, inplace = True)       # eliminamos columnas inservibles

data.head()
```

```python
data['BootsPack'].unique()
```

```python
valor = '-'
filas_coincidentes = data[data['PlayerNationality'] == valor].index          # filtrar por indice
data_seleccionado = data.loc[filas_coincidentes, ['PlayerNationality', 'PlayerName', 'BootsName', 'BootsBrand', 'BootsTopPlayers', 'PlayerPosition']]

data_seleccionado.head()
```

```python
mask = data.isnull()

null_counts = mask.sum()

print(null_counts)
```

```python
porcent_null = (data.isnull().sum(axis = 1) / data.shape[1]) * 100  # se suman los nulos por las filas y esa suma se divide entre el numero total de columnas del dataframe, para obtener el porcentaje de valores nulos en cada fila 
porcent_bool = porcent_null > 25        # si el valor de una fila es True, significa que mas del 25% de valores de esa fila son nulos
data = data.loc[~porcent_bool, :]       # seleccionamos con loc las filas que no tengan mas del 25% de nulos invirtiendo 'porcent_bool' con '~', y todas las columnas con ':'
```

```python
data.shape
```

```python
mask = data.isnull()

null_counts = mask.sum()

print(null_counts)
```

```python
data.tail()
```

```python
data.fillna(value = 'Unknown', inplace = True)      # rellenamos nulos con 'Unknown'

mask = data.isnull()
null_counts = mask.sum()
print(null_counts)
```

```python
data.tail()
```

```python
data.head()
```

```python
data['BootsBrand'] = data['BootsBrand'].replace('adidas', 'Adidas')     # arreglamos el nombre de 'Adidas'

data.head()
```

```python
data['ClubName'] = data['ClubName'].str.replace('S.P.A.L. 2013', 'S.P.A.L.')        # arreglamos el nombre del equipo S.P.A.L
```

```python
data['PlayerNationality'].unique()
```

```python
valor = '-'
filas_coincidentes = data[data['PlayerNationality'] == valor].index          # filtrar por indice
data_seleccionado = data.loc[filas_coincidentes, ['PlayerNationality', 'PlayerName', 'BootsName', 'BootsBrand', 'BootsTopPlayers', 'PlayerPosition']]

data_seleccionado.head()
```

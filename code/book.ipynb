```python
mask = data.isnull()

null_counts = mask.sum()

print(null_counts)
```

```python
data.drop(['Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17'], axis = 1, inplace = True)

data.head()
```

```python
data['BootsPack'].unique()
```

```python
valor = '-'
filas_coincidentes = data[data['PlayerNationality'] == valor].index          # filtrar por indice
data_seleccionado = data.loc[filas_coincidentes, ['PlayerNationality', 'PlayerName', 'BootsName', 'BootsBrand', 'BootsTopPlayers', 'PlayerPosition']]

data_seleccionado.head()
```

```python
mask = data.isnull()

null_counts = mask.sum()

print(null_counts)
```

```python
porcent_null = (data.isnull().sum(axis = 1) / data.shape[1]) * 100  # se suman los nulos por las filas y esa suma se divide entre el numero total de columnas del dataframe, para obtener el porcentaje de valores nulos en cada fila 
porcent_bool = porcent_null > 25        # si el valor de una fila es True, significa que mas del 75% de valores de esa fila son nulos
data = data.loc[~porcent_bool, :]       # seleccionamos con loc las filas que no tengan mas del 75% de nulos invirtiendo 'porcent_bool' con '~', y todas las columnas con ':'
```

```python
data.shape
```

```python
mask = data.isnull()

null_counts = mask.sum()

print(null_counts)
```

```python
data.tail()
```

```python
data.fillna(value = 'Unknown', inplace = True)

mask = data.isnull()
null_counts = mask.sum()
print(null_counts)
```

```python
data.tail()
```
